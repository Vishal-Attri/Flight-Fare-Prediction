## EDA
From description we can see that Date_of_Journey is a object data type,\ Therefore, we have to convert this datatype into timestamp so as to use this column properly for prediction.

For this we require pandas **to_datetime** to convert object data type to datetime dtype.

**dt.day** and **dt.month** method will be used seperately because they extract only day/month of that date

```python
train_data['Journey_date']=pd.to_datetime(train_data.Date_of_Journey,format="%d/%m/%Y").dt.day
```

```python
train_data['Journey_month']=pd.to_datetime(train_data["Date_of_Journey"],format='%d/%m/%Y').dt.month
```

```python
train_data['Journey_year']=pd.to_datetime(train_data["Date_of_Journey"],format='%d/%m/%Y').dt.year
```

```python
train_data.head()
```

```python
train_data.Journey_year.value_counts()
```

#### As all the values are from the same year we won't be needing this column

```python
train_data.drop(['Date_of_Journey','Journey_year'], axis= 1, inplace= True)
```

```python
train_data.head()
```

### Same process for **Departure time** and **Arrival time**

```python
train_data['Dep_hour']= pd.to_datetime(train_data.Dep_Time).dt.hour
train_data['Dep_min'] = pd.to_datetime(train_data.Dep_Time).dt.minute
```

```python
train_data['Arv_hour']= pd.to_datetime(train_data.Arrival_Time).dt.hour
train_data['Arv_min'] = pd.to_datetime(train_data.Arrival_Time).dt.minute
```

```python
train_data.drop(['Dep_Time','Arrival_Time'], axis=1, inplace= True)
```

```python
train_data.head()
```

#### Now for the duration column we will convert into a list and convert that data into integer

```python
duration=list(train_data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:               # checking if duration only contains hour or minutes then adding the missing
        if "h"in duration[i]:
            duration[i]=duration[i].strip() + "0m"        # adds 0 minutes
        else:
            duration[i]="0h" + duration[i]                # adds 0 hours

# making 2 new lists to add as cloumns in the dataset
duration_hours=[]
duration_mins=[]

for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep= 'h')[0]))                    # extracting hours from data
    duration_mins .append(int(duration[i].split(sep= "h")[-1].split(sep ="m")[0]))         # extracting minutes 
```

```python
train_data["Duration_hours"]=duration_hours
train_data["Duration_mins"]=duration_mins
```

```python
train_data.drop(['Duration'], axis=1, inplace =True)
```

### Handling Categorical Data
One can find many ways to handle categorical data. Some of them categorical data are,

1.**Nominal data** --> data are not in any order --> **OneHotEncoder** is used in this case

2.**Ordinal data** --> data are in order --> **LabelEncoder** is used in this case

# Hyperparameter Tuning
- Choose following method for hyperparameter tuning
  1. **RandomizedSearchCV** --> Fast
  2.  **GridSearchCV**
- Assign hyperparameters in form of dictionery
- Fit the model
- Check best paramters and best score

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in range(100, 1200, 100)]

# Number of features to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in range(5, 30, 6)]

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations

rf_random = RandomizedSearchCV(estimator = rf_reg, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state =5, n_jobs = -1)
```

```python
rf_random.fit(x_train,y_train)
```
